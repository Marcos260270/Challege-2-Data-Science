```python
import pandas as pd
import json

# Ruta al archivo (ajusta si es necesario)
file_path = "TelecomX_Data.json"

# Cargar los datos desde el JSON
with open(file_path, 'r') as f:
    data = json.load(f)

# Convertir a DataFrame
df = pd.json_normalize(data)

# Ver las primeras filas
#print(df.head())
df.head(10)
```

```python
#Convierte el campo account.Charges.Total, a numerico
import numpy as np
df['account.Charges.Total'] = df['account.Charges.Total'].replace(" ", np.nan)
df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')
```

```python
#seleccionar columnas a limpiar
columnas_a_limpieza = [
    'Churn',
    'customer.gender',
    'customer.Partner',
    'customer.Dependents',
    'phone.PhoneService',
    'internet.InternetService'
]

#Elimina registros con cadenas vacÃ­as o solo espacios en esas columnas

# Quitar espacios en blanco y filtrar filas vacÃ­as en esas columnas
for col in columnas_a_limpieza:
    df = df[df[col].str.strip() != '']

#Quitar tambiÃ©n los valores nulos (NaN)
df = df.dropna(subset=columnas_a_limpieza)

#Normaliza los strings (todo en minÃºsculas, sin espacios)
for col in columnas_a_limpieza:
    df[col] = df[col].str.strip().str.lower()


```

```python
#pie chart para ver evasores segÃºn monto total
# FunciÃ³n para mostrar % y cantidad
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar solo clientes que hicieron churn
df_churn = df[df['Churn'] == 'yes'].copy()

# Eliminar nulos (por si existen)
df_churn = df_churn[df_churn['account.Charges.Total'].notna()]

# Crear rangos de gasto total
bins = [0, 500, 1500, 3000, df_churn['account.Charges.Total'].max()]
labels = ['0-500', '501-1500', '1501-3000', '3001+']
df_churn['charges_rango'] = pd.cut(df_churn['account.Charges.Total'], bins=bins, labels=labels, include_lowest=True)

# Contar churn por rango de gasto
churn_por_cargo = df_churn['charges_rango'].value_counts().sort_index()

# FunciÃ³n para mostrar % y cantidad
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count})'
    return my_autopct

# Pie chart
plt.figure(figsize=(7,7))
plt.pie(
    churn_por_cargo,
    labels=churn_por_cargo.index,
    autopct=make_autopct(churn_por_cargo),
    startangle=90,
    colors=sns.color_palette('pastel')[0:len(churn_por_cargo)],
    explode=(0.05,) * len(churn_por_cargo)
)
plt.title('DistribuciÃ³n de Clientes que Hicieron Churn segÃºn Cargos Totales')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
